```python
import plotly.graph_objects as go
from geopy.geocoders import MapBox

# Replace 'YOUR_MAPBOX_ACCESS_TOKEN' with your actual Mapbox access token
mapbox_access_token = 'YOUR_MAPBOX_ACCESS_TOKEN'

# Initialize Mapbox Geocoding API
geolocator = MapBox(api_key=mapbox_access_token)

# Get user input for the location they want to visit
user_input = input("Enter the name of the place you want to visit: ")

# Perform geocoding to get coordinates
location = geolocator.geocode(user_input)

if location:
    # Get latitude and longitude
    latitude = location.latitude
    longitude = location.longitude
    
    # Create a Plotly scattermapbox figure
    fig = go.Figure(go.Scattermapbox(
        lat=[latitude],
        lon=[longitude],
        mode='markers',
        marker=go.scattermapbox.Marker(size=14),
        text=user_input,
    ))
    
    # Update layout to use Mapbox style and center the map
    fig.update_layout(
        hovermode='closest',
        mapbox=dict(
            accesstoken=mapbox_access_token,
            style='mapbox://styles/mapbox/streets-v11',  # Mapbox style
            center=dict(lon=longitude, lat=latitude),  # Center map around the location
            zoom=12,  # Adjust zoom level
        )
    )
    
    # Display the map
    fig.show()
else:
    print("Location not found. Please try again.")
```

```python
import requests
import json
import plotly.express as px

token = "pk.eyJ1IjoiZmdob3J5IiwiYSI6ImNscjR0ajM2dDE1dDAyaW84cHFrMnB5c2EifQ._-6eJJ519s5a1zkqYlQMXg"  # Replace with your Mapbox access token

def locations(search_texts: list):
    all_data = {'lat': [], 'lon': [], 'name':[]}

    for search_text in search_texts:
        url = f"https://api.mapbox.com/geocoding/v5/mapbox.places/{search_text}.json?proximity=ip"
        params = {'access_token': token}
        response = requests.get(url, params=params)
        data = json.loads(response.text)
        features = data.get('features', [])
        if features:
            long, lat = features[0].get('center')
            all_data['lat'].append(lat)
            all_data['lon'].append(long)
            all_data['name'].append(search_text)

    if all_data['lat'] and all_data['lon']:
        fig = px.scatter_mapbox(all_data, lat="lat", lon="lon",
                                zoom=10, height=400,width = 800, hover_name = "name"
                                )
        fig.update_traces(marker=dict(size=10, color='RoyalBlue', symbol='circle'))

        fig.update_layout(mapbox_style="satellite-streets",
                          mapbox_accesstoken=token)

        fig.show()

# Display multiple locations on the map
locations(["Eiffel Tower, Paris,France","Louvre Museum,Paris,France"])
```
